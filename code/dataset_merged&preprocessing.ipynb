```python
bus = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/project/3rd project_recommender_system/data_preparation/business.csv')
rv = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/project/3rd project_recommender_system/data_preparation/review.csv')
```

## Merge Dataset

```python
display(bus.info(),rv.info())
```

```python
# Merge the two datasets using the 'business_id' attribute
process = pd.merge(bus, rv, on='business_id')
# Perform additional data cleaning 
process.dropna(inplace=True)
process.head()
```

```python
process.info()
```

```python
process['attributes']
```

```python
# feature extraction
final = process.drop(['address', 'city', 'postal_code', 'longitude','latitude', 'review_count', 'stars_y', 'hours', 'review_id', 'date'], axis=1)
final.head()
# 'business_id', 'name', 'state', 'stars_x', 'is_open', 'attributes', 'categories', 'user_id', 'text', 'review_text', 'useful', 'funny', 'cool'
```
