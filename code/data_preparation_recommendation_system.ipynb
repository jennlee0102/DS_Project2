```python
# Determine the maximum number of unique values in a cell
max_unique_values = 10

# Create new columns for each unique value
for i in range(max_unique_values):
    bus[f"value_{i}"] = False

# Split the values in the original column and set the corresponding boolean values in the new columns
for i, row in bus.iterrows():
    if row['categories'] is None:
      continue
    values = row["categories"].split(",")
    for j in range(min(len(values), max_unique_values)):
        bus.loc[i, f"value_{j}"] = True

# Remove the original column
# bus = bus.drop("categories", axis=1)
```

## Review Dataset

```python
reviews = pd.read_json("/content/drive/MyDrive/Colab Notebooks/project/3rd project_recommender_system/original json data/yelp_academic_dataset_review.json",
                       lines=True, chunksize=100000)
# read the data 
for review in reviews:
    rv = review
    break

display(rv.head(2))
```

```python
rv.isnull().sum()
```

```python
rv['text'] = rv['text'].str.lower()
rv.head(5)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, SnowballStemmer
import string
import re, os

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

stemmer = SnowballStemmer('english')

def preprocess(text):
    # Tokenize the text
    tokens = nltk.word_tokenize(text)

    # Remove stopwords, punctuation marks, and non-alphanumeric characters, and stem the remaining words
    tokens = [stemmer.stem(re.sub(r'[^a-zA-Z]', '', token.lower())) for token in tokens if not token.lower() in nltk.corpus.stopwords.words('english') and not token.lower() in string.punctuation]

    return tokens

# Tokenize and stem the reviews
rv['review_tokens'] = rv['text'].apply(preprocess)

# Create a vocabulary of unique words
vocabulary = set()
for tokens in rv['review_tokens']:
    vocabulary.update(tokens)
vocabulary = sorted(list(vocabulary))
```
